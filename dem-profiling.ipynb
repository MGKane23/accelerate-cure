Drop ['BARCODE2_V2', 'SITE_ID', 'PATIENT_ID', 'DTBLDR  Date of blood draw', 'VISIT_ID', 'VISIT_ID.1', 'ROW_NO']
Rename remaining columns: disease, diag_dis, age, sex, ht,ht_units, wt, wt_units, dom_hand
Convert values: sex, ht_cm, wt_kg, dom_hand, diag_dis, disease
These columns need to be mapped from numbers to values
Drop unused columns: ['ht', 'ht_units', 'wt', 'wt_units']
Profile to check, then pickle to dem.pkl

```python
dem2 = demfirst.drop(columns=['BARCODE2_V2', 'SITE_ID', 'PATIENT_ID', 'DTBLDR  Date of blood draw', 'VISIT_ID', 'VISIT_ID.1', 'ROW_NO'])
```

```python
colnames = ['disease', 'diag_dis', 'age', 'sex', 'ht', 'ht_units', 'wt', 'wt_units', 'dom_hand']
dem2.columns = colnames
dem2
```

```python
disease = {1.0:'MS', 2.0:'NMO', 3.0:'ON', 4.0:'TM', 5.0:'ADEM', 6.0:'CIS', np.nan:'CONTROL'}
dem2['disease_mapped'] = dem2['disease'].map(disease)
diag = {1.0:'CONTROL', 2.0:'CASE'}
dem2['diag_mapped'] = dem2['diag_dis'].map(diag)
hand = {1.0:'RIGHT', 2.0:'LEFT', 3.0:'AMBIDEXTROUS', 4.0:'UNKNOWN'}
dem2['hand_mapped'] = dem2['dom_hand'].map(hand)
sex = {1.0:'MALE', 2.0:'FEMALE'}
dem2['sex_mapped'] = dem2['sex'].map(sex)
dem2
```

```python
# listcomprension here
# for w,u in zip(dem2['wt'], dem2['wt_units']):
#     if u == 1.0:
dem2['wt_kg'] = [(w * 0.45359237) if u == 1.0 else u for w,u in zip(dem2['wt'], dem2['wt_units'])]
dem2['ht_cm'] = [(h * 2.54) if u == 1.0 else u for h,u in zip(dem2['ht'], dem2['ht_units'])]
dem2
```

```python
dem3 = dem2.drop(columns=colnames
                )
dem3
```

```python
dem3.columns = ['disease', 'diag_dis', 'dom_hand', 'sex', 'wt_kg', 'ht_cm']
dem3
```

```python
dem = dem3
dem.to_pickle('pkl/dem.pkl')
```
